# A Method for Calibrating Probabilistic Forecasts

The fundamental concept behind this method is that, statistically speaking, forecast error consists of two parts: random error and systematic error. The calibration approach that follows determines both components of the statastical forecast error. The systematic error is corrected and knowledge of the random error is exploited to produce a probabilitic forecast.

What follows is an attempt to demonstrate the concepts behind this approach using randomly generated datasets.

-----

## Single Member

First, import the necessary modules to conduct the analysis


```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from scipy import ndimage
from mpl_toolkits.axes_grid1 import ImageGrid

%pylab inline

# Variables needed for analysis
samples = 1000000
lim = 40
cmap = plt.cm.Reds_r
```

Next, generate a random two-dimensional frequency distribution composite where observations fell relative to a forecast point for a deterministic forecast. The origin, point (0, 0), is considered to be the forecast. For this idealized case, the observation distribution is generated from a Gaussian distribution.

```python
class Member(object):
    def __init__(self, locx=0, locy=0, stdx=1, stdy=1, samples=1000, lim=20, step=1):
        
        # Configure the grid
        drange = range(-lim, lim+1, step)
        self.x, self.y = np.meshgrid(drange, drange)
        self.lim = lim
        self.step = step

        # Generate the random "observations"
        self.xpts = np.random.normal(locx, stdx, (samples,))
        self.ypts = np.random.normal(locy, stdy, (samples,))
        
        # Grid the observations
        z, xx, yy = np.histogram2d(self.xpts, self.ypts, bins=[drange, drange], 
                                   normed=False)
        self.data = z.T                             # Frequency Count
        self.pdata = self.data / self.data.sum()    # Probability

        # Create masked versions for plotting
        self.mdata = np.ma.array(self.data)
        self.mdata[self.data <= 0] = np.ma.masked
        self.pmdata = self.mdata / self.data.sum()

        # Center of Mass Calculations
        cxpt, cypt = ndimage.center_of_mass(self.data.T)
        self.cxpt = cxpt - lim
        self.cypt = cypt - lim
        self.cx = int(self.cxpt)
        self.cy = int(self.cypt)
```

Now, we need to use the class to generate our freqency distribution for our deterministic forecast.

```python
# Generate the distribution
mem1 = Member(locx=5, stdx=10, locy=5, stdy=10, lim=lim, samples=samples)
```

```python
# Set up the figure
fig = plt.figure(figsize=(12, 6))
grid = ImageGrid(fig, 111, nrows_ncols = (1, 2), axes_pad = 2, direction='row',
                add_all=True, share_all=True, aspect='equal', label_mode = "all",
                cbar_mode='each', cbar_location='right', cbar_pad=0.1, cbar_size='5%')
ax1 = grid[0]; cax1 = grid.cbar_axes[0]
ax2 = grid[1]; cax2 = grid.cbar_axes[1]


# Plot the frequency distribution
cbar1 = ax1.pcolormesh(mem1.x, mem1.y, mem1.mdata, cmap=cmap)
ax1.plot(0, 0, 'wo', markersize=10)
ax1.plot(mem1.cx, mem1.cy, 'k.', markersize=10)
ax1.axhline(0, lw=0.75, ls=':', color='k')
ax1.axvline(0, lw=0.75, ls=':', color='k')
ax1.set_xlim(-lim, lim)
ax1.set_ylim(-lim, lim)
ax1.set_title('Member 1 Frequency Distribution')
cax1.colorbar(cbar1)
print '\n\n'
print 'The total number of observations in the domain is:', mem1.data.sum()
print 'The maximum frequency is %i observations.' % (mem1.data.max())
print 'The center of mass of the frequency distribution is displaced (%f, %f).' % (mem1.cx, mem1.cy)
print 'The black dot is the center of mass of the frequency distribution: %i observations.' % (
        mem1.data[int(mem1.cx)+mem1.lim, int(mem1.cy)+mem1.lim])
print 'The white dot is the representative forecast.'


# Plot the probability distribution
cbar2 = ax2.pcolormesh(mem1.x, mem1.y, mem1.pmdata, cmap=cmap)
ax2.plot(0, 0, 'wo', markersize=10)
ax2.plot(mem1.cx, mem1.cy, 'k.', markersize=10)
ax2.axhline(0, lw=0.75, ls=':', color='k')
ax2.axvline(0, lw=0.75, ls=':', color='k')
ax2.set_xlim(-lim, lim)
ax2.set_ylim(-lim, lim)
ax2.set_title('Member 1 Probability Distribution')
cax2.colorbar(cbar2)
print '\n\n'
print 'The total probability in the domain is:', mem1.pdata.sum()
print 'The maximum probability is %f%%.' % (mem1.pdata.max() * 100.)
print 'The center of mass of the probability distribution is displaced (%f, %f).' % (mem1.cx, mem1.cy)
print 'The black dot is the center of mass of the probability distribution: %f%%.' % (
        mem1.pdata[int(mem1.cx)+mem1.lim, int(mem1.cy)+mem1.lim] * 100.)
print 'The white dot is the representative forecast.'


print '\n\n'
```

Using this visualization, it is readily apparent that there are random and systematic errors associated with the "forecast". The systematic error is given by the center of mass displacement vector. The random error is given by the observation frequency distribution. Using this information in a forecast sense is not extremely difficult. The general idea is to use the systematic error to "correct" the forecast by shifting the prediction to remove the systematic error. Once this is done, the determinstic yes/no forecast can be converted into a spatial probabilistic forecast by replacing the yes/no forecast with the spatial probability distribution -- which is based on the historical performace of the model.

For example, the following class generates the frequency and probabily distributions assuming there was no systematic error. This is achieved by taking a forecast grid point -- originally at point (0, 0) -- and "shifts" it by the displacement vector, resulting in a new, "corrected" forecast point (x, y) -- which is the now located at the center of mass of the previous distribution. (As mentioned before, one could use a displacement vector based on the maximum proability instead of the center of mass.)

```python
# Create a class to do the correction of individual members
class CorrectedMember(object):
    def __init__(self, mem):
        
        # Set up the grid
        self.x = mem.x
        self.y = mem.y
        self.lim = mem.lim
        self.step = mem.step

        # Correct the data
        self.data = ndimage.interpolation.shift(mem.data.T, (-mem.cx, -mem.cy), order=0).T
        self.pdata = self.data / self.data.sum()

        # Generate masked versions for plotting
        self.mdata = np.ma.array(self.data)
        self.mdata[self.data <= 0] = np.ma.masked
        self.pmdata = np.ma.array(self.pdata)
        self.pmdata[self.pdata <= 0] = np.ma.masked

        # Determine the new Center of Mass
        cxpt, cypt = ndimage.center_of_mass(self.data.T)
        self.cxpt = cxpt - self.lim
        self.cypt = cypt - self.lim
        self.cx = int(self.cxpt)
        self.cy = int(self.cypt)
```

```python
cmem1 = CorrectedMember(mem1)


# Set up the figure
fig = plt.figure(figsize=(12, 6))
grid = ImageGrid(fig, 111, nrows_ncols = (1, 2), axes_pad = 2, direction='row',
                add_all=True, share_all=True, aspect='equal', label_mode = "all",
                cbar_mode='each', cbar_location='right', cbar_pad=0.1, cbar_size='5%')
ax1 = grid[0]; cax1 = grid.cbar_axes[0]
ax2 = grid[1]; cax2 = grid.cbar_axes[1]


# Plot the frequency distribution
cbar1 = ax1.pcolormesh(cmem1.x, cmem1.y, cmem1.mdata, cmap=cmap)
ax1.plot(0, 0, 'wo', markersize=10)
ax1.plot(cmem1.cx, cmem1.cy, 'k.', markersize=10)
ax1.axhline(0, lw=0.75, ls=':', color='k')
ax1.axvline(0, lw=0.75, ls=':', color='k')
ax1.set_xlim(-lim, lim)
ax1.set_ylim(-lim, lim)
ax1.set_title('Member 1 Frequency Distribution')
cax1.colorbar(cbar1)
print '\n\n'
print 'The total number of observations in the domain is:', mem1.data.sum()
print 'The maximum frequency is %i observations.' % (mem1.data.max())
print 'The center of mass of the frequency distribution is displaced (%f, %f).' % (mem1.cx, mem1.cy)
print 'The black dot is the center of mass of the frequency distribution: %i observations.' % (
        mem1.data[int(mem1.cx)+mem1.lim, int(mem1.cy)+mem1.lim])
print 'The white dot is the representative forecast.'


# Plot the probability distribution
cbar2 = ax2.pcolormesh(cmem1.x, cmem1.y, cmem1.pmdata, cmap=cmap)
ax2.plot(0, 0, 'wo', markersize=10)
ax2.plot(cmem1.cx, cmem1.cy, 'k.', markersize=10)
ax2.axhline(0, lw=0.75, ls=':', color='k')
ax2.axvline(0, lw=0.75, ls=':', color='k')
ax2.set_xlim(-lim, lim)
ax2.set_ylim(-lim, lim)
ax2.set_title('Member 1 Probability Distribution')
cax2.colorbar(cbar2)
print '\n\n'
print 'The total probability in the domain is:', mem1.pdata.sum()
print 'The maximum probability is %f%%.' % (mem1.pdata.max() * 100.)
print 'The center of mass of the probability distribution is displaced (%f, %f).' % (mem1.cx, mem1.cy)
print 'The black dot is the center of mass of the probability distribution: %f%%.' % (
        mem1.pdata[int(mem1.cx)+mem1.lim, int(mem1.cy)+mem1.lim] * 100.)
print 'The white dot is the representative forecast.'


print '\n\n'
```

In principle the calibration method is simply:

    1. Find the model's systematic and random errors by utilizing a relative frequency distribution 
       of forecast points to observations.
    2. Everytime a model forecasts said event, shift it by the displacement vector. (Correct the 
       systematic error.)
    3. Replace the forecast point with the relative probability distribution. (Use the random error 
       to "smooth" the forecast.)
        
Previous results (Marsh et al. 2012) have shown that using this method results in generally reliable probabilitistic forecasts of preciptiation amounts exceeding 1" in 6 hours from a single model. Comparing probabilistic precipitation forecasts of exceeding 1" in 6 hours from a deterministic member to those generated from a 15 member ensemble have shown the forecasts from the deterministic member to be more reliable.

## Extending to an Ensemble

How to extend this method from a single member to an ensemble has been a source of debate amongst various particpants in the Hazardous Weather Testbed. Is it better to:

    1. Take the ensemble mean and then apply calibration method to the ensemble mean (Averaging then Correcting); or
    2. Apply the calibration method to each member of the ensemble and then take the mean of the corrected members.
       (Correcting then Averaging)

My hypotheses are:

    1. Each approach will produce reliable probabilistic forecasts as the underlying calibration
       method is designed to produce reliable forecasts.
    2. "Correcting then Averaging" will produce a "more narrow forecast" than "Averaging then Correcting"
       (i.e., it is "smoothed" over a smaller area). In the context of fitting gaussian functions to 
       the underlying probability distribution, a more narrow forecast equates to a sharper forecast.
    3. Given that both approaches are assumed to produce reliable forecasts, then a sharper forecast 
       is better.

The rest of this document provides an idealized justification of hypothesis 2.

-----

### "Average then Correct" Approach

To begin, first we must generate the relative frequency distributions for the members of our ensemble. For now we'll begin with a 2 member ensemble. We will also assume that each model has a bias of 1. That means each observation is associated with one and only one forecast point.




```python
# Generate the distributions
mem1 = Member(locx=10, stdx=7.5, locy=10, stdy=7.5, lim=lim, samples=samples)
mem2 = Member(locx=-10, stdx=5, locy=0, stdy=5, lim=lim, samples=samples)
```

Now, let's create a class to generate the ensemble of these members:

```python
# Create a class to generate an ensemble from a group of members
class Ensemble(object):
    def __init__(self, *args):
        
        # Ensemble the Members
        N = len(args)
        mem1 = args[0]
        combined = mem1.data.copy().astype('float')
        combined += mem1.data
        for arg in args[1:]:
            combined += arg.data
        combined /= N
        self.data = combined                      # frequency counts
        self.pdata = combined / combined.sum()    # probabilities

        # Set up the grid
        self.x = mem1.x
        self.y = mem1.y
        self.lim = mem1.lim
        self.step = mem1.step

        # Center of Mass Calculations
        cxpt, cypt = ndimage.center_of_mass(self.data.T)
        self.cxpt = (cxpt - self.lim)
        self.cypt = (cypt - self.lim)
        self.cx = int(self.cxpt)
        self.cy = int(self.cypt)
        
        # Create masked versions of the distributions for plotting
        self.mdata = np.ma.array(self.data)
        self.mdata[self.data <= 0] = np.ma.masked
        self.pmdata = np.ma.array(self.pdata)
        self.pmdata[self.pdata <= 0] = np.ma.masked
        
```

```python
# Ensemble the forecasts
ens = Ensemble(mem1, mem2)
```

### "Correct then Average" Approach

Now consider the case where each member of the ensemble is corrected before being ensembled. To do this, first we must produce the corrected members. (Remember, again, "corrected" here simply means that the systematic error is being removed.) After correcting each member, next we generate the "Correct then Average" ensemble.

The bottom set of figures are from the corrected members and thus the "correct then average" ensemble.
