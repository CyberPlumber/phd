```python
def plot_hist(h, ax, cax, norm=None, cmap=plt.cm.spectral, dx=None, title=None, ellipse=False, size=None):
    xlim = (h.shape[0] - 1) / 2
    ylim = (h.shape[1] - 1) / 2
    x = range(-xlim, xlim+1)
    y = range(-ylim, ylim+1)
    x, y = np.meshgrid(x, y)
    cbar = ax.pcolormesh(x, y, h, norm=norm, cmap=cmap)
    cax.colorbar(cbar)
    ax.plot(0, 0, 'ko')
    if title:
        if not size: size = 12
        ax.set_title(title, size=size)
    if dx:
        ax.set_yticklabels(ax.get_yticks().astype(int)*dx)
        ax.set_xticklabels(ax.get_xticks().astype(int)*dx)
    if ellipse:
        sigx, sigy, xrot = mp.kde.fit_gauss2d(h)
        ells1 = [Ellipse(xy=(0, 0), width=sigx*2*i, height=sigy*2*i, angle=xrot,
                 linewidth=1, fill=False, color='k', label="Sigma %i" % (i))
                 for i in np.arange(0.5, 2.1, 0.5)]
        [ax.add_artist(e) for e in ells1]

def find_mask(h, max_shift):
    h = data[0]
    xlim = (h.shape[0] - 1) / 2
    ylim = (h.shape[1] - 1) / 2
    x = range(h.shape[0])
    y = range(h.shape[1])
    x, y = np.meshgrid(x, y)
    gpoints = zip(x.ravel(), y.ravel())
    gtree = ss.KDTree(gpoints)
    pinds = gtree.query_ball_point((xlim, ylim), xlim-max_shift, p=2)
    pts = np.array(gpoints)[pinds]
    xx, yy = zip(*pts)
    tmp = np.zeros_like(x)
    for i,j in zip(xx, yy):
        tmp[i,j] = 1
    tmp = np.ma.asarray(tmp)
    tmp[tmp <= 0] = np.ma.masked
    return tmp.mask

def correct_hist(tdata, **kwargs):
    kind = kwargs.get('kind', 'center_of_mass')
    cxx = []; cyy = []
    for i in range(tdata.shape[0]):
        hr = ndimage.rotate(tdata[i].copy(), -90, order=0, prefilter=False)
        if kind.lower() == 'max_value':
            cy, cx = np.where(tdata[i] == tdata[i].max())
            cx = cx[0]; cy = cy[0]
        else:
            cy, cx = ndimage.center_of_mass(hr)
        cx -= xlim
        cy -= ylim
        cxx.append(cx)
        cyy.append(cy)
        tdata[i] = ndimage.shift(tdata[i].copy(), (-cy, -cx), order=0, prefilter=False)
    max_shift = np.ceil(max([max(cxx), max(cyy)])).astype(int)
    return tdata, cxx, cyy, max_shift
```

```python
data = []
for member in ssef:
    froot = os.path.join(thresh_root, member)
    fpath = os.path.join(froot, 'total.npz')
    f = np.load(fpath)
    data.append(f['total'])
data = np.ma.array(data)
data[data <= 0] = np.ma.masked
mdata = data.mean(axis=0)
xlim = (data[0].shape[0] - 1) / 2
ylim = (data[0].shape[1] - 1) / 2
```
